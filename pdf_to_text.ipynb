```python
import requests
import os
from rcare.readutils import DataFrameUtils as Utils
import sqlite3;
import pandas as pd;
```

```python
tika_url = "http://localhost:9998/tika"
```

```python
DATA_PATH = 'data/lexpredict-contraxsuite-samples/dan/';
DATA_FILES = list(filter(lambda x: x.endswith(".pdf"), os.listdir(DATA_PATH)))
src_dest = list(map(lambda x : ( DATA_PATH + x, DATA_PATH + Utils.join_list(x.split(" "), "_") + ".txt") , DATA_FILES))

for x in src_dest:
    #print (x[1])
    files = {'file': ('report.xls', open(x[0], 'rb'), 'application/pdf', {'Expires': '0'})}
    headers = {"Content-type": "application/pdf", "Accept": "text/plain"}
    r = requests.put(tika_url, files=files, headers=headers)
    #print (r.status_code, r.content)    
    if(r.status_code == requests.codes.ok):
        with open(x[1], "w") as f: 
            f.write(r.text) 
    else: 
        print (r.status_code, x[0])
        

#src_dest
```
