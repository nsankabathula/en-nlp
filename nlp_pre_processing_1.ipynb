```python
import logging;
logging.basicConfig(filename='nlp_pre_processing_1.log',format='%(module)s||%(funcName)s||[%(process)d:%(thread)d||%(asctime)s||%(levelname)s||%(message)s', level=logging.DEBUG,
                    datefmt=logging.Formatter.default_time_format)

import numpy as np
import pandas as pd
#from lexnlp.extract.rcare.models import Feature
from lexnlp.extract.rcare.features import get_features
#from lexnlp.extract.rcare.utils import Utils
from lexnlp.extract.rcare.readutils import DataFrameUtils as Utils
#Utils.DATA_PATH = 'Utils.DATA_PATH' # 'data/lexpredict-contraxsuite-samples/agreements/'
logging.info('TEST>>>');
Utils.DATA_DIRS
```

```python
# 1. Clean and write to file with : name: data. File name 001.txt
for d in Utils.DATA_DIRS:
    dir_path = Utils.DATA_PATH + d;    
    logging.info(dir_path)
    if(d == 'construction'):
        df = Utils.read_data(path = dir_path)
        df.to_csv(dir_path+ '/'+'001.txt',header = True, index  = False , doublequote = True, sep = '|', escapechar=' ');
```

```python
def get_f(text):
    f = get_features(text)        
    return pd.Series(f.to_dict()) 

def function_on_df(df):
    '''df[['regulation', 'amounts', 'citations', 'conditions', 'constraints', 
     'dates', 'definitions', 'durations', 'organizations', 'stfd_persons', 'nltk_persons', 'nltk_geopolitical', 
     'nltk_companies', 'nltk_parties', 'nltk_re_companies', 'address', 'titles']] = df.data.apply(get_f);
    '''
    df = df.join(df.data.apply(get_f))
    return df;

def process_df(df ):
    df_array = [];
    df_split = np.array_split(df, 2)    
    for x in df_split: 
        logging.info('Starting Processing %d', len(x.index));             
        try:
            df_i = x
            df_i = Utils.parallelize_dataframe(df_i, function_on_df);
            df_array.append(df_i)
        except Exception as err: 
            logging.exception(str(err));   
       
        
    logging.info('End of Processing.');
    return df_array;

#df_1 = Utils.parallelize_dataframe(df[:500], function_on_df)
def concat_df(df_array):
    logging.info('Starting Concat');
    df = pd.DataFrame();
    
    for dfitem in df_array:
        logging.info('Concat - df ');
        df = pd.concat([df, dfitem], ignore_index  = True)  
    
    return df;

def concat_file_data(dir_path):
    df = Utils.read_data(path = dir_path)
    df.to_csv(dir_path+ '001.txt',header = True, index  = False , doublequote = True, sep = '|', escapechar=' ');

def process_all(directory= None):
    for d in Utils.DATA_DIRS:
        dir_path = Utils.DATA_PATH + d + '/'; 
        
        if(directory == None or d == directory):             
            logging.info(dir_path)
            concat_file_data(dir_path)
            df = Utils.read_csv(path = dir_path, nrows = None);
            df_array = process_df(df)
            df = concat_df(df_array)
            df.to_csv(dir_path + '002.txt',header = True, index  = False, doublequote = True, sep = '|', escapechar=' ');
            if(d == directory ):
                return df;
```
