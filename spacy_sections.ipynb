```python
import spacy
from spacy.matcher import PhraseMatcher, Matcher
from spacy.tokens import Span, Doc, Token
import random

def tokens (doc):
    tokens = []
    for token in doc:
        tokens.append(token)
    return tokens

def get_sections (doc, matcher):
    doc = nlp(doc.text)
    
    print (matcher)
    matches = matcher(doc)
    prev_end = 0
    r_sections = [];
    print ('in sections', matches)
    for id, pos_s, pos_e in matches:
        print ('in matches')
        if(prev_end == 0):
            r_sections.append(doc[prev_end: pos_s])
        else:        
            r_sections.append(doc[prev_end: pos_s])
            print(prev_end, pos_s, pos_e)
        prev_end = pos_s    
    r_sections.append(doc[prev_end: ])
    return r_sections

Doc.set_extension('sections', force = True, method=get_sections)
Doc.set_extension('tokens', method=tokens, force = True)



class IMatcher(object):
    name = 'IMatcher'   

class TermMatcher(IMatcher):
    name = 'TermMatcher'

    def __init__(self, nlp, terms, label):
        patterns = [nlp(text) for text in terms]
        print (patterns)
        self.matcher = PhraseMatcher(nlp.vocab)
        self.matcher.add(label, None, *patterns)

    def __call__(self, doc):
        matches = self.matcher(doc)
        for match_id, start, end in matches:
            span = Span(doc, start, end, label=match_id)
            doc.ents = list(doc.ents) + [span]
        return doc
    
class PatternMatcher(IMatcher):
    name = 'PatternMatcher'

    def __init__(self, nlp, pattern, label):        
        self.matcher = Matcher(nlp.vocab)
        self.matcher.add(label, None, pattern)       

    def __call__(self, doc):
        matches = self.matcher(doc)
        for match_id, start, end in matches:
            span = Span(doc, start, end, label=match_id)
            doc.ents = list(doc.ents) + [span]
        return doc
    
    
nlp = spacy.load('en_core_web_sm')
terms = (u'cat', u'dog', u'tree kangaroo', u'giant sea spider')
phone_pattern = [{'ORTH': '('}, {'SHAPE': 'ddd'}, {'ORTH': ')'}, {'SHAPE': 'ddd'},
           {'ORTH': '-', 'OP': '?'}, {'SHAPE': 'ddd'}]            
            
          


term_entity_matcher = TermMatcher(nlp, terms, 'ANIMAL')
phone_entity_matcher = PatternMatcher(nlp, phone_pattern, 'PHONE_NUMBER')

#phone_entity_matcher1.name = 'test'
#print (phone_entity_matcher.name, phone_entity_matcher1.name)

nlp.add_pipe(term_entity_matcher, after='ner')
nlp.add_pipe(phone_entity_matcher, after='ner')

print(nlp.pipe_names)  # the components in the pipeline
doc = nlp(file_text)
#doc = nlp(u"This is a text about Barack Obama and a dog. Call me at (123) 456 789 or (123) 456 789 or +(123) 456 789")
#rint([(ent.text, ent.label_) for ent in doc.ents])
```

```python
def similarity_matrix(spac_docs):
    len_i = len(spac_docs)
    len_j = len(spac_docs)
    similarity_m = [.0] * len_i
    for i in range(len_i):
        doc_i = spac_docs[i]
        similarity_m[i] = [.0] * len_j
        for j in range(len_j):
            doc_j = spac_docs[j]
            similarity_m[i][j] = doc_i.similarity(doc_j)
    return similarity_m
df_con.head(3)
```
