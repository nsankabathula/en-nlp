```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn

# Display plots in this notebook, instead of externally. 
from pylab import rcParams
rcParams['figure.figsize'] = 16, 8
%matplotlib inline

DATA_PATH = 'data/lexpredict-contraxsuite-samples/credit_card_agreements/synchrony/' 
def get_files_in_path(path, extension = ".txt", black_list = [] ):
    filenames = list(filter(lambda x: (x not in black_list and \
                                       (x.endswith(extension) or x.endswith(extension.upper())  )) , \
                            os.listdir(path )))
    return filenames
#inaugural = [nlp(open(doc, errors='ignore').read()) for doc in inauguralFilenames]
```

```python
synchronyFileName = get_files_in_path(DATA_PATH)

synchrony = [s_nlp(open(DATA_PATH + doc, errors='ignore').read()) for doc in synchronyFileName]
```

```python
labels = synchronyFileName
sentLengths = [ np.mean([len(sent) for sent in doc.sents]) for doc in synchrony ]
pd.Series(sentLengths, index=labels).plot(kind='bar')
```

```python
similarities = [ [doc.similarity(other) for other in synchrony] for doc in synchrony ]
similaritiesDf = pd.DataFrame(similarities, columns=labels, index=labels)
```

```python
rcParams['figure.figsize'] = 16, 8
seaborn.heatmap(similaritiesDf)
```
